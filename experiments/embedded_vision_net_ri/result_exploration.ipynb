```python
import pandas as pd
import numpy as np
import pickle as pkl
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
metrics = pd.read_pickle(Path("metrics.pkl"))
```

```python
def minseries(series):
    s = []
    for i in series:
        if s:
            s.append(min(s[-1], i))
        else:
            s.append(i)
    return s
```

```python
metrics_ae = metrics[metrics['Task'] == "AE"]
metrics_rn = metrics[metrics['Task'] == "RANDOM"]
```

```python
plt.plot(minseries(metrics_rn["val_error"]), label="RN")
plt.plot(minseries(metrics_ae["val_error"]), label="AE")
plt.axvline(20, color="grey", linestyle="--")
plt.legend()
plt.show()
```
